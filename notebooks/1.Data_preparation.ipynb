There are different ways to treat a date. We extract month and encode it.

```python
df['Sow_Maize_month'] = df['Sow_Maize'].str.extract('([a-zA-Z]{3})')
df['Harvest_Maize_month'] = df['Harvest_Maize'].str.extract('([a-zA-Z]{3})')
```

```python
from calendar import month_abbr

lower_ma = [m.lower() for m in month_abbr]
df['Sow_Maize_month_int'] = df['Sow_Maize_month'].str.lower().map(lambda m: lower_ma.index(m)).astype('Int8')
df['Harvest_Maize_month_int'] = df['Harvest_Maize_month'].str.lower().map(lambda m: lower_ma.index(m)).astype('Int8')
df = df.drop(columns=['Sow_Maize', 'Harvest_Maize','Harvest_Maize_month','Sow_Maize_month'])
```

```python
df
```

#### Add time between sow and harvest

```python
# Find rows wher Harvest is lower then Sow, and assign value of columns Next_year to 1, otherwise 0
df.loc[:,'havest_next_year'] = (df['Harvest_Maize_month_int'] < df['Sow_Maize_month_int']).astype(int)
df['havest_next_year']


df.loc[:,'sow_to_harvest_months'] = df.loc[:,'Harvest_Maize_month_int'] - df.loc[:,'Sow_Maize_month_int'] + 12*df.loc[:,'havest_next_year']

# Drop column harvest_next_year
df = df.drop(columns=['havest_next_year'])

## Plot distribution of sow_to_harvest_months
plt.figure(figsize=(12, 4))  # Adjust the figsize as needed
sns.histplot(df['sow_to_harvest_months'])
plt.title('Time span between sowing and harvest in months')
plt.show()
```

#### Prepare weather data

```python
### Filter columns starting by pcp_, tmax, tmin and spi
pcp_columns = [col for col in df.columns if col.startswith('pcp_')].copy()
tmax_columns = [col for col in df.columns if col.startswith('tmax_')].copy()
tmin_columns = [col for col in df.columns if col.startswith('tmin_')].copy()
spi_columns = [col for col in df.columns if col.startswith('spi_')].copy()

## Calcualte mean of pcp columns
df['pcp_mean'] = df[pcp_columns].mean(axis=1)
df['tmax_mean'] = df[tmax_columns].mean(axis=1)
df['tmin_mean'] = df[tmin_columns].mean(axis=1)
df['spi_mean'] = df[spi_columns].mean(axis=1)

# Drop columns
df = df.drop(columns=pcp_columns)
df = df.drop(columns=tmax_columns)
df = df.drop(columns=tmin_columns)
df = df.drop(columns=spi_columns)

df[['pcp_mean','tmax_mean','tmin_mean','spi_mean']]
```

#### Calculate lagged variables

```python
# Creating 3 lags of the dependant variable (-1 year, -2 years, -3 years)
for i in range(1,4):
    df['maize_lag-'+str(i)] = df.sort_values(by=['Year']).groupby('Farm')['Y_maize_major'].shift(i)
    df['pcp_mean_lag-'+str(i)] = df.sort_values(by=['Year']).groupby('Farm')['pcp_mean'].shift(i)
    df['tmax_mean_lag-'+str(i)] = df.sort_values(by=['Year']).groupby('Farm')['tmax_mean'].shift(i)
    df['tmin_mean_lag-'+str(i)] = df.sort_values(by=['Year']).groupby('Farm')['tmin_mean'].shift(i)
    df['spi_mean_lag-'+str(i)] = df.sort_values(by=['Year']).groupby('Farm')['spi_mean'].shift(i)
```

```python
subset = ['maize_lag-1','maize_lag-2','maize_lag-3','pcp_mean_lag-1','tmax_mean_lag-1','tmin_mean_lag-1','spi_mean_lag-1']
df.loc[:, subset] = df.loc[:, subset].fillna(0) # Because some farms do not have past years data
df = df[df['Year']>2006] # Remove the 3 years kept for lagging
```
