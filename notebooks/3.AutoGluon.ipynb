```python
import pandas as pd

from autogluon.tabular import TabularDataset, TabularPredictor
```

## Data preparation

Read in data from prepared file.

```python
df = pd.read_csv('df_prepped.csv')
pd.set_option('display.max_columns', None)
df
```

For faster demo testing, create subsampling. Default is whole dataset.

```python
subsample_size = len(df)  # subsample subset of data for faster demo; full dataset (32K) fit takes appr 18 min, sample of 3K takes 1.5 min.
df = df.sample(n=subsample_size, random_state=0)
```

```python
df.shape
```

Additionally, `Countries` and `Farm` attributes as categorical are dropped.

```python
df = df.drop(['Countries','Farm'], axis=1)
df
```

Create train and test sets.

```python
# Separate a test set, the year 2016
df_test = df[df.Year == 2016]
df_train = df[df.Year < 2016]
```

```python
df_test.shape, df_train.shape
```

## AutoGluon: fit, predict, feature importance

Reference: https://auto.gluon.ai/stable/tutorials/tabular/tabular-essentials.html

Dependent variable for this regression task is `Y_maize_major`.

```python
label = 'Y_maize_major'
df_train[label].describe()
```

### Fit and predict on RMSE

Initialize and fit AutoGluon TabularPredictor. As two metrics - RMSE and MAE - are used in this project, these need to be fitted separately. Default metric for AutoGluon is RMSE.

```python
# takes about 18 minutes
metric = 'root_mean_squared_error' # rmse is default metric

predictor_rmse = TabularPredictor(label=label, eval_metric=metric).fit(df_train)
```

Now our trained model is used to make predictions on the new data:

```python
y_pred_rmse = predictor_rmse.predict(df_test.drop(columns=[label]))
y_pred_rmse.head()
```

Next, we can evaluate the predictor on the (labeled) test data:

```python
predictor_rmse.evaluate(df_test, silent=True)
```

For evaluation metrics where higher values are worse (like RMSE), AutoGluon will flip their sign and print them as negative values during training (as it internally assumes higher values are better).

Each model is evaluated individually, in this case based on RMSE.

```python
predictor_rmse.leaderboard(df_test)
```

Best model is ExtraTreesMSE with RMSE on test set of 0.3069.

For the whole model, RMSE on test data is 0.3315 and MAE 0.2133.

### Fit and predict on MAE

Same procedure is now followed by using mean absolute error as metric. First fit on train data:

```python
metric = 'mean_absolute_error'
predictor_mae = TabularPredictor(label=label, eval_metric=metric).fit(df_train)
```

Then make predictions:

```python
y_pred_mae = predictor_mae.predict(df_test.drop(columns=[label]))
y_pred_mae.head()
```

Then evaluation on labeled test data:

```python
predictor_mae.evaluate(df_test, silent=True)
```

And then model ranking based on MAE:

```python
predictor_mae.leaderboard(df_test)
```

Best model based on MAE is also ExtraTreesMSE with MAE on test data of 0.1901.

### Feature importance

To better understand our trained predictor, we can estimate the overall importance of each feature via `feature_importance()`. First, model fitted based on RMSE:

```python
predictor_rmse.feature_importance(df_test) #14 minutes
```

Then, model fitted based on MAE:

```python
predictor_mae.feature_importance(df_test)
```
